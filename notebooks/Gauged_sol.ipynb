x' = -\frac{2}{r}x+ \frac{1}{(1-\lambda g^{2} \sin^{2} \alpha)} \left[ 2\lambda g x y \sin^{2}\alpha  -\frac{g^{2}}{2}\sin 2\alpha (1-\lambda x^{2}) +4 m^{2} \sin \alpha \right]; \,\,\\
g'=y; \,\,\\
y'= -\frac{2}{r}y+ 4K g \sin^{2}\alpha (1+\lambda x^{2}) $
