```python
plt.plot(T2.rn, T2.α, label = 'α0 = 2.5, ω = 0.8, λ = 0.0073')
plt.plot(T3.rn, T3.α, label = 'α0 = 6,   ω = 1, λ = 0.00073')
plt.plot(T1.rn, T1.α, label = 'α0 = 7,   ω = 0.4, λ = 0.0073')
#plt.plot(Q5.rn, Q5.α, label = 'λ = 0.00073')
plt.text(101, 3.14,'$π$')
plt.text(101, 9.4,'$3π$')
#plt.xlim([-2, 107])
#plt.ylim([-0, 2.0])
plt.xlabel('$r$', fontsize=22)
plt.title('α(r) for ω < 2 m = 2')
plt.legend(loc='best')
#plt.show()
plt.savefig('a_ung.png', dpi =1000)
```
